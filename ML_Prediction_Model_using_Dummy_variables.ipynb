```python
## Creating prediction model using dummy variable

#importing relevant libraries and reading csv file

import pandas as pd
df = pd.read_csv('C:\\Users\\user\\Documents\\PythonPractiseFiles\\home_price_wih_dummies.csv')
df

```

```python
##geting dummy variables for column town 

dummies = pd.get_dummies(df.town)
dummies 
```

```python

#merged the two files to get one
merged= pd.concat([df,dummies],axis='columns')
merged
```

```python
#we can drop one out of three dummy variables to prevent Dummy Variable Trap
#also droping town column coz now its repeated

final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
#creating Regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#Making two variable X and Y (X with independent variables and Y with Dependent)

X=final.drop('price',axis='columns')
X
```

```python

Y = final.price
Y
```

```python
#fitting the model with X and Y
model.fit(X,Y)
```

```python
#Predicting for values
model.predict([[2800,0,1]])
```

```python
#predicting for other values
model.predict([[3400,0,0]])
```

```python
#checking the accuracy of model
model.score(X,Y)
```
